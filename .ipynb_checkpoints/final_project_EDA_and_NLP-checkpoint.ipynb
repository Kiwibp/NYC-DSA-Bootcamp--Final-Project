```python
def punct_space(token):
    """
    helper function to eliminate tokens
    that are pure punctuation or whitespace
    """
    
    return token.is_punct or token.is_space
            
def lemmatized_sentence_corpus(filename, column):
    """
    generator function to use spaCy to parse reviews,
    lemmatize the text, and yield sentences
    """
    if column == 'Title':
        for parsed_titles in nlp.pipe(all_items_df_nlp['Title'].astype('unicode').values):        
            for sent in parsed_titles.sents:
                yield u' '.join([token.lemma_ for token in sent
                                 if not punct_space(token)])
    elif column == 'Description':        
        for parsed_descriptions in nlp.pipe(all_items_df_nlp['Description'].astype('unicode').values):        
            for sent in parsed_descriptions.sents:
                yield u' '.join([token.lemma_ for token in sent
                                 if not punct_space(token)])
    else:
        print('Please specify Title or Description column.')
```

```python
def trigram_bow_generator(filepath):
    """
    generator function to read reviews from a file
    and yield a bag-of-words representation
    """
    
    for title in LineSentence(filepath):
        yield trigram_dictionary.doc2bow(title)
```

### Titles

```python
trigram_normalized_text = LineSentence('trigram_normalized_titles.txt')

# learn the dictionary by iterating over all of the reviews
trigram_dictionary = Dictionary(trigram_normalized_text)

# filter tokens that are very rare or too common from
# the dictionary (filter_extremes) and reassign integer ids (compactify)
# trigram_dictionary.filter_extremes(no_below=10, no_above=0.4)
# trigram_dictionary.compactify()

# trigram_dictionary.save(trigram_dictionary_filepath)
    
# # load the finished dictionary from disk
# trigram_dictionary = Dictionary.load(trigram_dictionary_filepath)
```

```python
# generate bag-of-words representations for
# all reviews and save them as a matrix
MmCorpus.serialize('trigram_normalized_titles.txt',
                   trigram_bow_generator('trigram_normalized_titles.txt'))
    
# load the finished bag-of-words corpus from disk
trigram_bow_corpus = MmCorpus('trigram_normalized_titles.txt')
```

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore')

    # workers => sets the parallelism, and should be
    # set to your number of physical cores minus one
    lda = LdaMulticore(trigram_bow_corpus,
                       num_topics=20,
                       id2word=trigram_dictionary)
    
# lda = LdaMulticore(trigram_bow_corpus,
#                        num_topics=50,
#                        id2word=trigram_dictionary,
#                        workers=3)
```

```python
def explore_topic(topic_number, topn=25):
    """
    accept a user-supplied topic number and
    print out a formatted list of the top terms
    """
        
    print(u'{:20} {}'.format(u'term', u'frequency') + u'\n')

    for term, frequency in lda.show_topic(topic_number):
        print(u'{:20} {:.3f}'.format(term, round(frequency, 3)))
```

```python
explore_topic(topic_number=9)
```

```python
LDAvis_prepared = pyLDAvis.gensim.prepare(lda, trigram_bow_corpus,
                                          trigram_dictionary)

with open('LDAvis_data.txt', 'wb') as f:
    pickle.dump(LDAvis_prepared, f)
        
# load the pre-prepared pyLDAvis data from disk
with open('LDAvis_data.txt', 'rb') as f:
    LDAvis_prepared = pickle.load(f)
```

```python
from collections import Counter
from itertools import chain

def get_vocab(lst):
    vocabcount_titles = Counter()
    vocabcount_desc = Counter()
    
    vocabcount = Counter()
    
    for title, desc in lst:
        if title == None or desc == None:
            continue
        else:
            vocabcount.update(w for w in title)
            vocabcount_titles.update(w for w in title)
            vocabcount_desc.update(w for w in desc)
            
            vocabcount.update(w for w in title)
            vocabcount.update(w for w in desc)
            
    vocab_titles = map(lambda x: x[0], sorted(vocabcount_titles.items(), key=lambda x: -x[1]))     
    vocab_desc = map(lambda x: x[0], sorted(vocabcount_desc.items(), key=lambda x: -x[1]))
    
    vocab = list(map(lambda x: x[0], sorted(vocabcount.items(), key=lambda x: -x[1])))
    
    return vocabcount_titles, vocab_titles, vocabcount_desc, vocab_desc, vocabcount, vocab
```

```python
vocabcount_titles, vocab_titles, vocabcount_desc, vocab_desc, vocabcount, vocab = get_vocab(combined_tokenized)
```
