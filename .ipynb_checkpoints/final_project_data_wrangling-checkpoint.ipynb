```python
import json
import pandas as pd
import numpy as np
import re
pd.options.display.max_colwidth = 500
```

```python
def load_json_files_into_list(file):
    #loading raw json files into lists 
    used_items_raw = []
    with open(file) as json_file:
        if "Facebook" in file:
            for line in json_file:
                used_items_raw.append(json.loads(line))
        elif "Letgo" in file:
            for line in json_file:
                used_items_raw.append(json.loads(line))
        else:
            for line in json_file:
                used_items_raw.append(json.loads(line))            
        return used_items_raw
    json_file.close()
    

fb_used_items_raw = load_json_files_into_list('Facebook_items.json')

print(len(fb_used_items_raw))

lg_used_items_raw = load_json_files_into_list("Letgo_items.json")

print(len(lg_used_items_raw))

cl_used_items_raw = load_json_files_into_list("Craigslist_items.json")

print(len(cl_used_items_raw))
```

```python
lg_used_items_raw[0]
```

```python
cl_used_items_raw[0]
```

```python
# Creating subset of keys for FB items
wanted_keys_fb = ['Title', 'Location', 'Price', 'Description', 'Url'] 

# creating subset of keys for LG and CL items
wanted_keys = wanted_keys_fb + ['Latitude', 'Longitude']

def clean_up_list_of_dicts(raw_list):
    '''
    This file cleans up a raw list of items and writes them to a text file.
    '''
    
    if raw_list == fb_used_items_raw:
        with open('fb_used_items_filtered.txt', 'w+') as file:
            for item in raw_list:
                for k in wanted_keys_fb:
                    lines = item[k].split()
                    split_lines = [line.replace('\n', '').strip() for line in lines]
                    file.write("{}".format(' '.join(split_lines)) + '\t')
                file.write('\n')
            
    elif raw_list == lg_used_items_raw:
        with open('lg_used_items_filtered.txt', 'w+') as file:
            for item in raw_list:
                for k in wanted_keys:
                    lines = str(item[k]).split()
                    split_lines = [line for line in lines if line.strip()]            
                    file.write("{}".format(' '.join(split_lines)) + '\t')
                file.write('\n')
                
    else:
        with open('cl_used_items_filtered.txt', 'w+') as file:
            for item in raw_list:
                for k in wanted_keys:
                    if k == 'Description':
                        lines = str(''.join(item[k])).split()
                        split_lines = [line.replace('\n', '').strip() for line in lines]
                        split_lines = ' '.join(split_lines)
                        file.write(split_lines + '\t')
                    else:
                        lines = str(item[k]).split()
                        split_lines = [line.replace('\n', '').strip() for line in lines]
                        file.write("{}".format(' '.join(split_lines)) + '\t')
                file.write('\n')
    
    file.close()  
```

```python
clean_up_list_of_dicts(fb_used_items_raw)
clean_up_list_of_dicts(lg_used_items_raw)
clean_up_list_of_dicts(cl_used_items_raw)
```

```python
#creating functions to adjust locations. FB df doesn't require any adjustments.
def adjust_location_values(locations_to_change, df):
    if df == 'lg':
        for location in locations_to_change:
            indexes = lg_items_df[lg_items_df['Location'].str.match(re.escape(location[0]))].index
            for index in indexes:
                lg_items_df.iloc[index, 1] = location[1]
        
        lg_items_df['Location'] = lg_items_df['Location'].str.title()
        lg_items_df['Location'] = lg_items_df['Location'].str[:-2] + lg_items_df['Location'].str[-2:].str.upper()
        
    
    else:
        for location in locations_to_change:
            indexes = cl_items_df[cl_items_df['Location'].str.match(re.escape(location[0]))].index
            for index in indexes:
                cl_items_df.iloc[index, 1] = location[1]
        
        cl_items_df['Location'] = cl_items_df['Location'].str.title()
        cl_items_df['Location'].where(cl_items_df['Location'] == 'Unspecified', 
                                      cl_items_df['Location'].str[:-2] + cl_items_df['Location'].str[-2:].str.upper(),
                                     inplace=True)
    
                
def drop_location_values(locations_to_drop, df):
    if df == 'lg':
        for location in locations_to_drop:
            indexes = lg_items_df[lg_items_df['Location'].str.match(re.escape(location))].index
            for index in indexes:
                lg_items_df.drop(index, inplace=True)
            lg_items_df.reset_index(inplace=True, drop=True)
    else:
        for location in locations_to_drop:
            indexes = cl_items_df[cl_items_df['Location'].str.match(re.escape(location))].index
            for index in indexes:
                cl_items_df.drop(index, inplace=True)
            cl_items_df.reset_index(inplace=True, drop=True)
```
